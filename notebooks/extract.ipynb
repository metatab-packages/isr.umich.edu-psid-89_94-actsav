```python
import seaborn as sns
import metapack as mp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display 

%matplotlib inline
sns.set_context('notebook')
mp.jupyter.init()
```

```python
#pkg = mp.jupyter.open_package()
pkg = mp.jupyter.open_source_package()
pkg
```

```python
dp = pkg.reference('source').resolved_url.fspath
```

```python
cs = [ (int(e['start']),int(e['end'])+1) for  e in pkg.resource('act94').columns()]
col_names = [ e['name'] for  e in pkg.resource('act94').columns()]

act94 = pd.read_fwf(dp.joinpath('ACT94.txt'), colspecs=cs)
act94.columns = col_names

act94['act94v16'] = pd.to_numeric(act94.act94v16.replace({'0-':'0', '3000-':'3000'}))

act94.head()
```
